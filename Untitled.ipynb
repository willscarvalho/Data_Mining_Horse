<h2> Atribute Information </h2>


1. Sugery:
* Yes, it had surgery. **(Sim, teve cirurgia.)**
* It was treated without sugery. **(Foi tratado sem cirurgia.)**

2. Age: 
* Adult horse. **(Cavalo adulto.)**
* Young. **(< 6  meses.)**

3. Hospital Number 
* Numeric Id. **(Id numérico.)** 
* the case number assigned to the horse (may not be unique if the horse is treated > 1 time). **(o número do caso atribuído ao cavalo (pode não ser único se o cavalo for tratado> 1 vez).)**

4. Rectal Temperature
* Linear. **(Linear.)**
* In  degrees celsius. **(Em graus celsius.)**
* An elevated temp may be reduced when the animal is in late shock. **(Uma temperatura elevada pode ser reduzida quando o animal está em choque tardio)**
* Normal temp is 37,8. **(A temperatura normal é 37,8.)**
* This parameter will usually change as the problem progresses, eg. may start out normal, then become elevated because of the lesion, passing back through the normal range as the horse goes into shock. **(este parâmetro geralmente muda conforme o problema progride, por exemplo: pode começar normal, então se tornar elevado por causa da lesão, voltando à faixa normal quando o cavalo entra em choque.)**

5. Pulse
* Linear. **(Linear)**
* The heart rate in beats per minute. **(A frequência cardíaca em batimentos por minuto.)**
* Is a reflection of the heart condition: 30 -40 is normal for adults. **(É um reflexo da condição cardíaca: 30 -40 é normal para adultos.)**
* rare to have a lower than normal rate although athletic horses may have a rate of 20-25. **(Raro ter uma taxa menor do que o normal, embora cavalos atléticos possam ter uma taxa de 20-25.)**
* Animals with painful lesions or suffering from circulatory shock may have an elevated heart rate. **(Animais com lesões dolorosas ou sofrendo de choque circulatório podem ter uma frequência cardíaca elevada.)**

6. Respiratory Rate
* Linear. **(Linear.)**
* Normal rate is 8 to 10. **(A taxa normal é de 8 a 10.)**
* Usefulness is doubtful due to the great fluctuations. **(Utilidade é duvidosa devido às grandes flutuações.)**

7. Temperature of Extremities
* _A subjective indication of peripheral circulation. **(Uma indicação subjetiva de circulação periférica.)**_
* _Possible values:_
* 1 = Normal. **(Normal.)** 
* 2 = Warm. **(Calorosa.)**
* 3 = Cool. **(Legal.)**
* 4 = Cold. **(Fria.)**
* _Cool to cold extremities indicate possible shock. **(Extremidades de frio a frio indicam possível choque.)**_
* _Hot extremities should correlate with an elevated rectal temp. **(as extremidades quentes devem se correlacionar com uma temperatura retal elevada.)**_

8. Peripheral Pulse 
* _Subjective. **(Subjetivo. Os valores possíveis são:)_**
* _Possible values are:_
* 1 = normal. **(Normal.)**
* 2 = increased. **(Aumentado.)**
* 3 = reduced. **(Reduzido.)**
* 4 = absent. **(Ausente.)**
* _Normal or increased p.p. (Peripheral Pulse) are indicative of adequate circulation while reduced or absent indicate poor perfusion. **(Normal ou aumento de p.p. (Pulso periférico) são indicativos de circulação adequada, enquanto a redução ou ausência indica perfusão insuficiente.)**_

9. Mucous Membranes 
* _A subjective measurement of colour. **(Uma medida subjetiva de cor.)**_
* _The possible values are:_
* 1 = normal pink. **(Rosa normal.)**
* 2 = bright pink. **(Rosa brilhante.)**
* 3 = pale pink. **(Rosa pálido.)**
* 4 = pale cyanotic. **(Cianótico pálido.)**
* 5 = bright red / injected. **(Vermelho brilhante / injetado.)**
* 6 = dark cyanotic. **(Cianótico escuro.)**
* _1 and 2 probably indicate a normal or slightly increased circulation. **(1 e 2 provavelmente indicam uma circulação normal ou ligeiramente aumentada.)**_
* _3 may occur in early shock. **(3 pode ocorrer no início do choque.)**_
* _4 and 6 are indicative of serious circulatory compromise. **(4 e 6 são indicativos de comprometimento circulatório sério.)**_
* _5 is more indicative of a septicemia. **(5 é mais indicativo de septicemia.)**

10.  Capillary Refill Time
* _A clinical judgement. The longer the refill, the poorer the circulation. **(Um julgamento clínico. Quanto mais tempo o refil, pior a circulação.)**_
* _The possible values:_
* 1 = < 3 seconds. **(1 = <3 segundos.)**
* 2 = >= 3 seconds. **(2 => = 3 segundos.)**

11. Pain, a subjective judgement of the horse's pain level

* 1 = alert, no pain. **(1 = alerta, sem dor.)**
* 2 = depressed. **(deprimido.)**
* 3 = intermittent mild pain. **(dor leve intermitente.)**
* 4 = intermittent severe pain. **(dor intensa intermitente.)**

* _In general, the more painful, the more likely it is to require surgery. **(Em geral, quanto mais dolorido, maior a probabilidade de exigir cirurgia.)**_
* _Prior treatment of pain may mask the pain level to some extent. **(O tratamento anterior da dor pode mascarar o nível de dor até certo ponto.)**_

12.  Peristalsis
* _An indication of the activity in the horse's gut. As the gut becomes more distended or the horse becomes more toxic, the activity decreases. **(Uma indicação da atividade no intestino do cavalo. Conforme o intestino se torna mais distendido ou o cavalo torna-se mais tóxico, a atividade diminui.)**_

13. Abdominal Distension

* _Possible values:_
* 1 = none. **(Nenhum.)**

14. Nasogastric Tube

* _Possible values:_
* 1 = None. **(Nenhum.)**_

<h3> Pré-processamento </h3>

```python
# Join datasets train and test

# save id
train_idx = dftrain.shape[0]
test_idx = dftest.shape[0]

# join train and test 
df_train_test = pd.concat(objs=[dftrain, dftest], axis=0).reset_index(drop=True)

# before
print("train.shape: ({} x {})".format(dftrain.shape[0], dftrain.shape[1]))
print("test.shape: ({} x {})\n".format(dftest.shape[0], dftest.shape[1]))

# after join
print("df_train_test: ({} x {})".format(df_train_test.shape[0], df_train_test.shape[1]))
```

```python
def Removing_Filling(df_train_test):
    for col in df_train_test.columns.values:
        if (pd.isna(df_train_test[col]).sum()) > 0:
            if pd.isna(df_train_test[col]).sum() > (50/100 * len(df_train_test)): 
                print(col,"removido") 
                df_train_test = df_train_test.drop([col], axis=1)
            elif (df_train_test[col].dtype == 'object'):
                df_train_test[col] = df_train_test[col].fillna(df_train_test[col].mode()[0])  
            else:
                df_train_test[col] = df_train_test[col].fillna(df_train_test[col].median())
    return df_train_test
```

```python
df_train_test = Removing_Filling(df_train_test)
```

```python
print(df_train_test.shape)
print(df_train_test.isna().sum())
```

```python
df_train_test["outcome"] = df_train_test["outcome"].astype('category').cat.codes
df_train_test.head()
```

```python
df_train_test_corr = df_train_test.corr()
corr_values = df_train_test_corr["outcome"].sort_values(ascending=False)
corr_values = abs(corr_values).sort_values(ascending=False)

print("Correlated")
print(abs(corr_values).sort_values(ascending=False))
```

```python
df_train_test = df_train_test.drop(columns=['hospital_number'], axis=1)
df_train_test = df_train_test.drop(columns=['respiratory_rate'], axis=1)
df_train_test = df_train_test.drop(columns=['lesion_3'], axis=1)
df_train_test = df_train_test.drop(columns=['rectal_temp'], axis=1)

df_train_test.head()
```

```python
df_train_test = pd.get_dummies(df_train_test)
df_train_test.head(10)
```

```python
# Recovering datasets train and test
dftrain = df_train_test.iloc[:train_idx]
dftest = df_train_test.iloc[train_idx:]

# Checking shape  
print("--- Dataset Train ---")
print("Variáveis:\t{}\nEntradas:\t{}\n".format(dftrain.shape[1], dftrain.shape[0]))

print("--- Dataset Test ---")
print("Variáveis:\t{}\nEntradas:\t{}".format(dftest.shape[1], dftest.shape[0]))
```

```python
X_train = dftrain.drop("outcome", axis=1).values
Y_train = dftrain["outcome"]
X_test  = dftest.drop("outcome", axis=1).values
Y_test  = dftest["outcome"]
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)


Y_pred1 = knn.predict(X_test)

acc_knn = round(knn.score(X_train, Y_train) * 100, 2)
print('Acurácia do modelo KNeighborsClassifier:',acc_knn, "\n")

# Confusion Matrix 
print(pd.crosstab(Y_test,Y_pred1, rownames=["Real"], colnames=["Predict"], margins=True))
```

```python
# Decision Tree
decision_tree = DecisionTreeClassifier(max_depth = 3)
decision_tree.fit(X_train, Y_train)

Y_pred2 = decision_tree.predict(X_test)

acc_decision_tree = round(decision_tree.score(X_train, Y_train) * 100, 2)
print('Acurácia do modelo DecisionTreeClassifier:',acc_decision_tree, "\n")

# Confusion Matrix 
print(pd.crosstab(Y_test,Y_pred2, rownames=["Real"], colnames=["Predict"], margins=True))
```

```python
# SVM

support_vm = svm.SVC()
support_vm.fit(X_train, Y_train)

Y_pred3 = support_vm.predict(X_test)

acc_svm = round(support_vm.score(X_train, Y_train) * 100, 2)
print('Acurácia do modelo Support Vector Machine:',acc_svm, "\n")

# Confusion Matrix 
print(pd.crosstab(Y_test,Y_pred3, rownames=["Real"], colnames=["Predict"], margins=True))
```

```python
# Gaussian Naive Bayes

gnb = GaussianNB()
gnb.fit(X_train, Y_train)

Y_pred4 = gnb.predict(X_test)

acc_gnb = round(gnb.score(X_train, Y_train) * 100, 2)
print('Acurácia do modelo Support Vector Machine:',acc_gnb, "\n")

# Confusion Matrix 
print(pd.crosstab(Y_test,Y_pred4, rownames=["Real"], colnames=["Predict"], margins=True))
```

```python
# MLP

mlp = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
mlp.fit(X_train, Y_train)

Y_pred5 = mlp.predict(X_test)
acc_mlp = round(mlp.score(X_train, Y_train) * 100, 2)
print('Acurácia do modelo Artificial Neural Network MLP:',acc_mlp, "\n")

# Confusion Matrix 
print(pd.crosstab(Y_test,Y_pred5, rownames=["Real"], colnames=["Predict"], margins=True))
```

```python
# Final Ranking 
results = pd.DataFrame({
    'Model': ['KNN', 'SVM', 'GaussianNB', 'ANN-MLP'],
    'Score': [acc_knn, acc_svm, acc_gnb, acc_mlp]})
    
result_df = results.sort_values(by='Score', ascending=False)
result_df = result_df.set_index('Score')
result_df.head(9)
```
